# Slice and dice the Datasets

```python
# Downloading and unzipping the datasets
!wget "https://archive.ics.uci.edu/ml/machine-learning-databases/00462/drugsCom_raw.zip"
!unzip drugsCom_raw.zip
```

```python
# loading a datasets
data_files = {"train":"drugsComTrain_raw.tsv",
              "test": "drugsComTest_raw.tsv"}
```

```python
drug_dataset = load_dataset("csv", data_files=data_files, delimiter="\t")
```

```python
drung_sample_dataset = drug_dataset["train"].shuffle(seed=42).select(range(1000))            # taking 1000 random data by shuffling from training part
drung_sample_dataset[:5]
```

```python
# checking whether the unnamed : 0 is a unique id for each patent or not
for split in drug_dataset.keys():
    assert len(drug_dataset[split]) == len(drug_dataset[split].unique("Unnamed: 0"))                     # check whether the Unnamed: 0 is unqiue or not
```

```python
# since the unnamed:0 is unique lets rename it by patient_id
drug_dataset = drug_dataset.rename_column(original_column_name="Unnamed: 0", new_column_name="patient_id")             # renaming column
drug_dataset
```

```python
# lowercasing every value in condition key


drug_dataset = drug_dataset.filter(lambda x: x["condition"] is not None)         # removing the none vales first
def normalize_condition(examples):
  return {'condition' : examples['condition'].lower()}                           # function to  lowercase value of condition key in datasets

drug_dataset = drug_dataset.map(normalize_condition)
```

```python
# creating a new column named as review length
def review_len(examples):
  return {"review_length": len(examples['review'].split())}

drug_dataset = drug_dataset.map(review_len)
```

```python
# checking whether the review lenght worked or not
drug_dataset["train"][0]
```

```python
# removing the dataset which has review length less than 30
drug_dataset = drug_dataset.filter(lambda x : x["review_length"] > 30)                                    # keeps the review with length greater than 30
```

```python
drug_dataset["train"][0]
```

```python
# Removing html character code in our review
import html
drug_dataset = drug_dataset.map(lambda x: {"review": html.unescape(x["review"])})                   # removing html code from review
```

```python
# Removing html character code using batch funciton in map
new_drug_dataset = drug_dataset.map(
    lambda x: {"review": [html.unescape(o) for o in x["review"]]}, batched=True          # batch return the list of value instead of value
)
```
